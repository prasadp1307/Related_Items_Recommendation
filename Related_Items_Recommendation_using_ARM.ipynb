#In Association Rule Mining Data is Standardize by using Transaction Encoder

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
A = pd.DataFrame(te.fit_transform(A),columns=te.columns_)
```

```python
def confidence(A,x,y):
    num = A[(A[x] == True) & (A[y] == True)].shape[0]
    den = A[A[x] == True].shape[0]
    conf = num/den
    return conf
```

```python
confidence(A,"Jam","Milk")
```

```python
confidence(A,"Jam","Brownbread")
```

```python
confidence(A,"Apple","Banans")
```

```python
for i in A.columns:
    for j in A.columns:
        print(i,j,supp(A,i,j),confidence(A,i,j))
```

```python
rule = []
support = []
conf = []
for i in A.columns:
    for j in A.columns:
        if(i !=j):
            rule.append(i + " " + j)
            support.append(supp(A,i,j))
            conf.append(confidence(A,i,j))
```

```python
T = pd.DataFrame([rule,support,conf]).T
T.columns = ["Rule","Supp","Conf"]
```
